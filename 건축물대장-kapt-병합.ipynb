# kapt 정제

```python
#'시도'가 서울인 데이터만 필터링
kapt_df = kapt_df[kapt_df['시도'].fillna('').str.startswith('서울')]
print(f'서울시 kapt : {len(kapt_df)}')
```

```python
# 도로명주소키

def make_key(s):
    return (
        s.astype('string')
         .str.replace(r'\s+', '', regex=True)
         .str.replace(r'\(.*?\)', '', regex=True)
         .str.strip()
)
    
kapt_df['도로명주소키'] = make_key(kapt_df['도로명주소'])
```

```python
# 법정동주소키

# ','뒤의 값 모두 제거
kapt_df['법정동주소키'] = (
    kapt_df['법정동주소']
        .astype('string')
        .str.replace(r',.*$', '', regex=True)
        .str.strip()
)

def clean_법정동주소키(addr):
    if pd.isna(addr):
        return addr
    
    # 공백 제거
    addr = re.sub(r'\s+', '', addr)
    
    # 건물명 제거
    match = re.search(r'(.*?(동|리|가)\d+[-]?\d*)', addr)
    if match:
        result = match.group(1)
        result = re.sub(r'-$', '', result)
        return result
    return addr

kapt_df['법정동주소키'] = kapt_df['법정동주소키'].apply(clean_법정동주소키)
```

```python
# 도로명주소키
건축물대장_df['도로명주소키'] = make_key(건축물대장_df['도로명대지위치'])
```

```python
# 법정동주소키
건축물대장_df['법정동주소키'] = (
    건축물대장_df['대지위치']
        .astype('string')
        .str.replace('번지', '', regex=False)
        .str.replace(r'\s+', '', regex=True)
)
```
