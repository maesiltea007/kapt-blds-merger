# kapt 정제

```python
#'시도'가 서울인 데이터만 필터링
kapt_df = kapt_df[kapt_df['시도'].fillna('').str.startswith('서울')]
print(f'kapt_df : {len(kapt_df)}')
```

```python
# 도로명주소키

def make_key(s):
    return (
        s.astype('string')
         .str.replace(r'\s+', '', regex=True)
         .str.replace(r'\(.*?\)', '', regex=True)
         .str.strip()
)
    
kapt_df['도로명주소키'] = make_key(kapt_df['도로명주소'])
```

```python
# 법정동주소키

# ','뒤의 값 모두 제거
kapt_df['법정동주소키'] = (
    kapt_df['법정동주소']
        .astype('string')
        .str.replace(r',.*$', '', regex=True)
        .str.strip()
)

def clean_법정동주소키(addr):
    if pd.isna(addr):
        return addr
    
    # 공백 제거
    addr = re.sub(r'\s+', '', addr)
    
    # 건물명 제거
    match = re.search(r'(.*?(동|리|가)\d+[-]?\d*)', addr)
    if match:
        result = match.group(1)
        result = re.sub(r'-$', '', result)
        return result
    return addr

kapt_df['법정동주소키'] = kapt_df['법정동주소키'].apply(clean_법정동주소키)
```

# 건축물대장 정제

```python
# 도로명주소키
건축물대장_df['도로명주소키'] = make_key(건축물대장_df['도로명대지위치'])
```

```python
# 법정동주소키
건축물대장_df['법정동주소키'] = (
    건축물대장_df['대지위치']
        .astype('string')
        .str.replace('번지', '', regex=False)
        .str.replace(r'\s+', '', regex=True)
)
```

```python
총괄_건축물대장_df = 건축물대장_df[건축물대장_df['그룹 종류'] == 0] # 총괄표제부만
print(f'건축물대장_df : {len(건축물대장_df)}')
print(f'총괄_건축물대장_df : {len(총괄_건축물대장_df)}')
```

# 병합

```python
# 1) 도로명주소키 기준으로 유일한 데이터만 선별
kapt_도로명주소_counts = kapt_df['도로명주소키'].value_counts(dropna=False)
총괄_건축물대장_도로명주소_counts  = 총괄_건축물대장_df['도로명주소키'].value_counts(dropna=False)
건축물대장_도로명주소_counts = 건축물대장_df['도로명주소키'].value_counts(dropna=False)

kapt_도로명주소_unique_mask = kapt_df['도로명주소키'].map(kapt_도로명주소_counts) == 1
총괄_건축물대장_도로명주소_unique_mask  = 총괄_건축물대장_df['도로명주소키'].map(총괄_건축물대장_도로명주소_counts) == 1
건축물대장_도로명주소_unique_mask  = 건축물대장_df['도로명주소키'].map(건축물대장_도로명주소_counts) == 1

kapt_with_unique_도로명주소 = kapt_df[kapt_도로명주소_unique_mask]
총괄_건축물대장_with_unique_도로명주소  = 총괄_건축물대장_df[총괄_건축물대장_도로명주소_unique_mask]
건축물대장_with_unique_도로명주소  = 건축물대장_df[건축물대장_도로명주소_unique_mask]

# 2) 도로명주소키로 join
joined_df_on_도로명주소키 = kapt_with_unique_도로명주소.merge(
    총괄_건축물대장_with_unique_도로명주소,
    on='도로명주소키',
    how='inner',
    suffixes=('_k_apt_도로명주소병합', '_총괄_건축물대장_도로명주소병합')
)
print(f'도로명주소키로 병합됨 : {len(joined_df_on_도로명주소키)}')

# 3) 도로명주소키 매칭에 실패한 데이터들
used_도로명주소키 = joined_df_on_도로명주소키['도로명주소키'].unique()
kapt_rest = kapt_df[~kapt_df['도로명주소키'].isin(used_도로명주소키)]
총괄_건축물대장_rest  = 총괄_건축물대장_df[~총괄_건축물대장_df['도로명주소키'].isin(used_도로명주소키)]

# 법정동주소키 기준으로 유일한 데이터만 선별
kapt_rest_법정동주소_counts = kapt_rest['법정동주소키'].value_counts(dropna=False)
총괄_건축물대장_rest_법정동주소_counts = 총괄_건축물대장_rest['법정동주소키'].value_counts(dropna=False)

kapt_rest_unique_mask = kapt_rest['법정동주소키'].map(kapt_rest_법정동주소_counts) == 1
총괄_건축물대장_rest_unique_mask = 총괄_건축물대장_rest['법정동주소키'].map(kapt_rest_법정동주소_counts) == 1

kapt_rest = kapt_rest[kapt_rest_unique_mask]
총괄_건축물대장_rest = 총괄_건축물대장_rest[총괄_건축물대장_rest_unique_mask]

# 4) 법정동주소키로 join
joined_df_on_법정동주소키 = kapt_rest.merge(
    총괄_건축물대장_rest,
    on='법정동주소키',
    how='inner',
    suffixes=('_k_apt_법정동주소병합', '_총괄_건축물대장_법정동주소병합')
)
print(f'법정동주소키로 병합됨: {len(joined_df_on_법정동주소키)}')

# 5) 법정동주소키 매칭에도 실패한 데이터들
used_법정동주소키 = joined_df_on_법정동주소키['법정동주소키'].unique()
kapt_rest = kapt_rest[~kapt_rest['법정동주소키'].isin(used_법정동주소키)]

# 6) 매칭에 실패한 데이터와 건축물대장 전체 병합
joined_df_with_건축물대장_on_도로명주소키 = kapt_rest.merge(
    건축물대장_with_unique_도로명주소,
    on='도로명주소키',
    how='inner',
    suffixes=('_k_apt_도로명주소병합', '_건축물대장_도로명주소병합')
)
print(f'건축물대장과 도로명주소키로 병합됨: {len(joined_df_with_건축물대장_on_도로명주소키)}')

used_도로명주소키_with_건축물대장 = joined_df_with_건축물대장_on_도로명주소키['도로명주소키'].unique()
kapt_rest = kapt_rest[~kapt_df['도로명주소키'].isin(used_도로명주소키_with_건축물대장)]
건축물대장_rest = 건축물대장_df[~건축물대장_df['도로명주소키'].isin(used_도로명주소키_with_건축물대장)]
# 건축물대장에서 법정동주소키 기준으로 유일한 키만 선별
건축물대장_rest_법정동주소키_counts = 건축물대장_rest['법정동주소키'].value_counts(dropna=False)
건축물대장_rest_법정동주소키_unique_mask = 건축물대장_rest['법정동주소키'].map(건축물대장_rest_법정동주소키_counts) == 1
건축물대장_rest = 건축물대장_rest[건축물대장_rest_법정동주소키_unique_mask]

joined_df_with_건축물대장_on_법정동주소키 = kapt_rest.merge(
    건축물대장_rest,
    on='법정동주소키',
    how='inner',
    suffixes=('_k_apt_법정동주소병합', '_총괄_건축물대장_법정동주소병합')
)
print(f'건축물대장과 법정동주소키로 병합됨: {len(joined_df_with_건축물대장_on_법정동주소키)}')


# 6) 최종 결합
joined_df = pd.concat([
    joined_df_on_도로명주소키, 
    joined_df_on_법정동주소키, 
    joined_df_with_건축물대장_on_도로명주소키, 
    joined_df_with_건축물대장_on_법정동주소키
    ], ignore_index=True)
print(f'병합된 결과 : {len(joined_df)}')
```
