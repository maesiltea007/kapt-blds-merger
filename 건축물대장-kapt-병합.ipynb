불러오기

```python
kapt_df = pd.read_excel(
    kapt_path,
)
print(f'len(kapt_df) : {len(kapt_df)}')
```

```python
건축물대장_df = pd.read_excel(
    건축물대장_path,
)
print(f'len(건축물대장_df) : {len(건축물대장_df)}')
```

# kapt 정제

```python
#'시도'가 서울인 데이터만 필터링
kapt_df = kapt_df[kapt_df['시도'].fillna('').str.startswith('서울')]
print(f'서울시 kapt : {len(kapt_df)}')
```

# 건축물대장 정제

```python
#총괄표제부인 것만
건축물대장apt_df = 건축물대장_df[건축물대장_df['그룹 종류'] == 0]
print(f'총괄표제부 건축물대장 : {len(건축물대장apt_df)}')
```
